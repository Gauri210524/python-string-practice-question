**Question - 1:**
Take a string input from the user and:

Print its length

Print its first character

Print its last character

***Hint:***
Use len() and indexing.

```python
#write your code here
text = input("Enter a string: ")

print("Length:", len(text))
print("First character:", text[0])
print("Last character:", text[-1])
```

**Question - 2:**
Given a string, print:

The string in lowercase

The string in uppercase

The string in title case

***Hint:***
Use string methods.

```python
#write your code here
text = input("Enter a string: ")

print("Lowercase:", text.lower())
print("Uppercase:", text.upper())
print("Title case:", text.title())
```

**Question - 3:**
Given a string with extra spaces at the beginning and end, remove the spaces and print the cleaned string.

Example:

text = "       data analyst   "


Expected Output: "data analyst"

***Hint:***
Use strip().

```python
#write your code here
text = " data analyst "

clean_text = text.strip()
print(clean_text)

```

**Question - 4:**
Given a string, count how many times the letter 'a' appears in it.

Example:

text = "data analysis"


Expected Output: 4

***Hint:***
Use loop or count().

```python
#write your code here
text = "data analysis"

count_a = text.count('a')
print(count_a)

```

**Question - 5:**
Given an email ID, extract and print the domain name.

Example:

email = "student@gmail.com"


Expected Output:

gmail.com


***Hint:***
Use split().

```python
#write your code here
email = "student@gmail.com"

domain = email.split('@')[1]

print(domain)
```
